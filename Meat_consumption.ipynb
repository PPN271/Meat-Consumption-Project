```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Example CO2 emissions data (Gt CO2/year)
# data from 1. Global Greenhouse Gas Emissions Data
data = {
    'Source': ['Electricity and Heat Production ', 'Industry', 'human-caused greenhouse gas emissions', 'Transportation', 'Buildings ', 'Other Energy'],
    'CO2 Emissions': [23, 24, 14.5, 15, 6, 17.5],
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.pie(df['CO2 Emissions'], labels=df['Source'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Global Emissions by Economic Sector')
plt.savefig('output/figure_1')
plt.show()
```

Agriculture, Forestry, and Other Land Use (22% of 2010 global greenhouse gas emissions): Greenhouse gas emissions from this sector come mostly from agriculture (cultivation of crops and livestock) and deforestation. This estimate does not include the CO2 that ecosystems remove from the atmosphere by sequestering carbon in biomass, dead organic matter, and soils, which offset approximately 20% of emissions from this sector.

```python
#data from Fact check: 2. How bad is eating meat for the climate?
# Carbon emissions data (percentages)

data = {
    'Source': ['Meat Consumption', 'Other Emissions'],
    'Percentage': [68, 32],
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.pie(df['Percentage'], labels=df['Source'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Carbon Emissions in Agriculture, Forestry, and Other Land Use Sector')
plt.savefig('output/figure_2')
plt.show()
```

```python
meat_types = ['Chicken', 'Pork', 'Beef', 'Seafood', 'Beyond Meat', 'Tofu']
amounts = [480 * 7 * 0.65, 700, 600 * 7 * 0.1, 400 * 7 * 0.05, 25, 200]
carbon_footprints = [10, 12, 99, 7.5, 2, 3]

# Calculate CO2 emissions for each type of meat and plant-based alternatives
co2_emissions = [amount * footprint for amount, footprint in zip(amounts, carbon_footprints)]

plt.figure(figsize=(10, 6))
bars = plt.bar(meat_types, co2_emissions)
plt.xlabel('Meat and Plant-based Alternatives')
plt.ylabel('CO2 Emissions (kg/week)')
plt.title('Weekly CO2 Emissions from Meat and Plant-based Alternatives at the Cafeteria')

# Add the values on top of each bar
for bar, value in zip(bars, co2_emissions):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, round(value, 2), ha='center', va='bottom')
plt.savefig('output/Weekly CO2 Emissions')
plt.show()
```
