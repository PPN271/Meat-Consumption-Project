```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Example CO2 emissions data (Gt CO2/year)
# data from 1. Global Greenhouse Gas Emissions Data
data = {
    'Source': ['Electricity and Heat Production ', 'Industry', 'Agriculture, Forestry, and Other Land Use', 'Transportation', 'Buildings ', 'Other Energy'],
    'CO2 Emissions': [23, 24, 22, 15, 6, 10],
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.pie(df['CO2 Emissions'], labels=df['Source'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Global Emissions by Economic Sector')
plt.show()
```

Agriculture, Forestry, and Other Land Use (22% of 2010 global greenhouse gas emissions): Greenhouse gas emissions from this sector come mostly from agriculture (cultivation of crops and livestock) and deforestation. This estimate does not include the CO2 that ecosystems remove from the atmosphere by sequestering carbon in biomass, dead organic matter, and soils, which offset approximately 20% of emissions from this sector.

```python
#data from Fact check: 2. How bad is eating meat for the climate?
# Carbon emissions data (percentages)

data = {
    'Source': ['Meat Consumption', 'Other Emissions'],
    'Percentage': [68, 32],
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
plt.pie(df['Percentage'], labels=df['Source'], autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Carbon Emissions in Agriculture, Forestry, and Other Land Use Sector')
plt.show()
```

According to FAO data, 14.5% of all human-caused greenhouse gas emissions are attributable to livestock farming, an industry that emits not only carbon dioxide (CO2), but also methane (CH4) and nitrous oxide (N2O) — two gases considered to play a similar role to CO2 in driving global warming. 

```python
# Carbon footprint data (kg CO2e/kg)
# data from 3. How sustainable are fake meats?
data = {
    'Food': ['Beef(beef herd)', 'Beef, dairy herd', 'Pork', 'Poultry meats', 'Plant-based', 'Tofu'],
    'Carbon Footprint': [99, 22, 11.5, 7.5, 2, 2.5],
}

df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))
sns.barplot(x='Food', y='Carbon Footprint', data=df)
plt.title('Carbon Footprint of Meat and Plant-based Alternatives')
plt.ylabel('Carbon Footprint (kg CO2eq)')
plt.show()
```
